Creating a figure with a grid of subplots is a very common task. So matplot lib includes plt.subplots, that creates a new figure and returns a NumPy array containing the created subplot objects 

```python
fig, axis = plt.subplots(2, 3)
axis
```

The axis array can be indexed like a 2D array. for example, axes[0, 1]. You can also indicate subplots should use the same x, y axis. using sharex and sharey, respectively.

## Adjusting the Spacing Around Subplots  
By defualt matlplotlib leaves a certain amount of padding around the outside of the subplots and spacing between plots. 
this can be resized using subplots_adjust method on Figure objects. 

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5) 
plt.subplots_adjust(wspace=0, hspace=0) #hspace and wspace ontrols the percent of the figure width and figure height.
```

## Ticks, Labels and Legends
The pyplot interface consists of methods such as xlim, sticks and xticklabels. Thes control the plot range, tick locations, and tick labels respectively. They can be used in two ways. 
* Called with no arguments returns the current parameter values 
* called with parameters sets the parameter value 

```python
fig = plt.figure()
ax = fig.add_subplot(1,1, 1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
# The rotation option sets the x tick labels at a 30-degree rotation
# set_xlabel gives a name to the x-axis and set_title the subplot title
ax.set_title('My first matplotlib plot')
ax.set_xlabel('stages')
# Modifying the y axis is a similar process 
props = {
    'title': 'my first matplotlib plot',
    'xlabel': 'Stages'
}
ax.set(**props)
```

## Adding lengends

```python
from numpy.random import randn  
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best')
```

# Annotationas and Drawing on a Subplot   
in addition to plot types you may whish to draw plot annotations e.g. arrows, text and annotate. below is a plot of the 2007 crash 


```python
from datetime import datetime 
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure() #create the figure 
ax = fig.add_subplot(1,1,1) #add one plot to the figure 
t = np.arange(0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
ax.plot(s)
ax.annotate('Amplitude', xy=(200, 1), xytext=(250, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
ax.set_ylim([-1.7, 2.00])
ax.set_title('Sine wave')
```
